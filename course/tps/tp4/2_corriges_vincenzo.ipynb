# Exercice 

- En générant des données aléatoires dans le langage de votre choix, reproduire le schéma Northwind en respectant les types (composé uniquement des tables `categories`, `products` et `suppliers`, on ne s'intéresse pas aux autres tables).

```python
import psycopg2

# Se connecter à la base de données
conn = psycopg2.connect(
    host="localhost",
    database="northwind_fake",
    user="user",
    password="password"
)

# Créer un curseur
cur = conn.cursor()

# Créer la table categories
cur.execute('''
    CREATE TABLE categories (
        category_id SERIAL PRIMARY KEY,
        category_name VARCHAR(255) NOT NULL,
        description TEXT
    )
''')

# Créer la table suppliers
cur.execute('''
    CREATE TABLE suppliers (
        supplier_id SERIAL PRIMARY KEY,
        supplier_name VARCHAR(255) NOT NULL,
        contact_name VARCHAR(255),
        contact_title VARCHAR(255),
        address VARCHAR(255),
        city VARCHAR(255),
        region VARCHAR(255),
        postal_code VARCHAR(20),
        country VARCHAR(255),
        phone VARCHAR(20),
        fax VARCHAR(20),
        homepage TEXT
    )
''')

# Créer la table products
cur.execute('''
    CREATE TABLE products (
        product_id SERIAL PRIMARY KEY,
        product_name VARCHAR(255) NOT NULL,
        category_id INT REFERENCES categories(category_id),
        supplier_id INT REFERENCES suppliers(supplier_id),
        unit_price DECIMAL(10, 2) NOT NULL,
        units_in_stock INT,
        description TEXT
    )
''')

# Valider et fermer la connexion
conn.commit()
cur.close()
conn.close()
```

```python
from faker import Faker
import random
import psycopg2

# Se connecter à la base de données
conn = psycopg2.connect(
    host="localhost",
    database="northwind_fake",
    user="user",
    password="password"
)

# Créer un curseur
cur = conn.cursor()

# Générer des données aléatoires
fake = Faker()

# Générer des catégories
for _ in range(5):
    cur.execute('''
        INSERT INTO categories (category_name, description)
        VALUES (%s, %s)
    ''', (fake.word(), fake.text()))

# Générer des fournisseurs
for _ in range(5):
    cur.execute('''
        INSERT INTO suppliers (supplier_name, contact_name, contact_title, address, city, region, postal_code, country, phone, fax, homepage)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    ''', (fake.company(), fake.name(), fake.job(), fake.street_address(), fake.city(), fake.state(), fake.zipcode()[:20], fake.country(), fake.phone_number()[:20], fake.phone_number()[:20], fake.url()))

# Générer des produits
for _ in range(20):
    cur.execute('''
        INSERT INTO products (product_name, category_id, supplier_id, unit_price, units_in_stock, description)
        VALUES (%s, %s, %s, %s, %s, %s)
    ''', (fake.word(), random.randint(1, 5), random.randint(1, 5), round(random.uniform(10, 100), 2), random.randint(10, 100), fake.text()))

# Valider et fermer la connexion
conn.commit()
cur.close()
conn.close()
```
